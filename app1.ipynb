```python
import autocorrect
import emoji
import contractions
from unidecode import unidecode
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

chat = {
    "brb": "be right back",
    "btw": "by the way",
    "lol": "laugh out loud",
    "omg": "oh my god",
    "ttyl": "talk to you later",
    "idk": "I don't know",
    "imo": "in my opinion",
    "imho": "in my humble opinion",
    "fyi": "for your information",
    "smh": "shaking my head",
    "np": "no problem",
    "tbh": "to be honest",
    "wbu": "what about you",
    "bc": "because",
    "afaik": "as far as I know",
    "asap": "as soon as possible",
    "atm": "at the moment",
    "bbl": "be back later",
    "bfn": "bye for now",
    "bff": "best friends forever",
    "cu": "see you",
    "cya": "see you",
    "dm": "direct message",
    "fb": "Facebook",
    "ftw": "for the win",
    "gg": "good game",
    "gr8": "great",
    "gtg": "got to go",
    "hbu": "how about you",
    "i'll":'i will',
    "ily": "I love you",
    "jk": "just kidding",
    "lmao": "laughing my ass off",
    "lmk": "let me know",
    "nvm": "never mind",
    "omw": "on my way",
    "pls": "please",
    "ppl": "people",
    "rofl": "rolling on the floor laughing",
    "thx": "thanks",
    "u": "you",
    "ur": "your",
    "yolo": "you only live once",
    "yw": "you're welcome",
    "ty": "thank you",
    "abt": "about"
}
def Cleaning(text): 
    text = text.lower() #1. lower
    ob = autocorrect.Speller() #2. autocorrect
    text = ob(text) #3. emojis
    text = emoji.demojize(text).replace(':',"") 
    text= contractions.fix(text) #4. contractions
    text= unidecode(text) #5. unicoding-- engilish accents
    res = []
    for i in text.split():
        if i in chat.keys():
            res.append(chat[i])
        else:
            res.append(i)   #6. chatwords cleaning
    text = " ".join(res) 
    text = re.sub("<.*?>","",text) #7.html tags
    text= re.sub(r'https\S+', '', text) #8.urls cleaning
    text= re.sub(r'[a-zA-Z0-9.%+-]+@[a-zA-Z0-9.-]+.[a-zA-Z]{2,}', '', text) #9.gmail cleaning
    text = re.sub(r"[^a-z A-Z ]","",text) #10. remove punctuations and digits
    # advanced cleaning
    preprocess = []
    for word in text.split():
        if word not in stopwords.words('english'):
            preprocess.append(word)

    text= ' '.join(preprocess)
    text= word_tokenize(text) 
    lemma  = WordNetLemmatizer()
    res=[]
    for i in text:
        res.append(lemma.lemmatize(i))
    text= ' '.join(res)
    return text
```

```python
df["question"] = df["question"].apply(Cleaning)
```

```python
df["text"] = df["text"].apply(Cleaning)
```
