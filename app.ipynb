```python
question = []
text = []
tags = []
for i in range(1,4010):
    url = f"https://stackoverflow.com/questions?tab=newest&page={i}"
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"}
    page = requests.get(url)
    soup = BeautifulSoup(page.text)
    for post in soup.find_all("div", class_="s-post-summary"):  # Adjust container class if needed
        # Extract question title (or np.nan if missing)
        title = post.find("h3", class_="s-post-summary--content-title")
        question.append(title.text.strip() if title else np.nan)

        # Extract text excerpt (or np.nan if missing)
        excerpt = post.find("div", class_="s-post-summary--content-excerpt")
        text.append(excerpt.text.strip() if excerpt else np.nan)

        # Extract tags (or np.nan if missing)
        tag_list = post.find("ul", class_="ml0 list-ls-none js-post-tag-list-wrapper d-inline")
        if tag_list:
            tags.append(",".join(tag.text for tag in tag_list.find_all("li")))
        else:
            tags.append(np.nan)
```

```python
len(tags)
```
